## Age wise Debt 

```python
total_median_debt = pd.DataFrame(data.loc[4:10,['Characteristic', 'Total Debt', 'Secured Debt', 'Unsecured Debt']]).reset_index(drop=True)
total_median_debt
```

```python
median_different_debt =  pd.DataFrame(data.loc[4:10,['Characteristic', 'Unnamed: 3','Unnamed: 4', 'Unnamed: 5', 'Unnamed: 7','Unnamed: 8', 'Unnamed: 9']].rename(index=str, columns={
    "Unnamed: 3": "Secure Home Debt", 
    "Unnamed: 4": "Secure Business Debt", 
    "Unnamed: 5": "Secure Vehicle Debt",
    "Unnamed: 7": "Unsecure Credit Card Debt", 
    "Unnamed: 8": "Unsecure Loans", 
    "Unnamed: 9":"Unsecure Other Debt"
})).reset_index(drop=True)

median_different_debt
```

```python
#final_df = pd.concat([total_mean_debt, unsecure_debt, secure_debt], axis=1)
age_median_debt = pd.merge(median_different_debt,total_median_debt, on=['Characteristic'])
age_median_debt
```

```python
file_name1 = "Agewise_median_of_debt2011.xlsx"
age_median_debt.to_excel(file_name1)
```

## Region wise Debt 

```python
a = data.loc[:,['Characteristic', 'Total Debt', 'Secured Debt', 'Unsecured Debt']]
region_df = pd.DataFrame(a[12:16]).reset_index(drop=True)
region_df
```

```python
b = data.loc[:,['Characteristic', 'Unnamed: 3','Unnamed: 4', 'Unnamed: 5', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9']]
secure_unsecure_debt_region = pd.DataFrame(b[12:16].rename(index=str, columns={
    "Unnamed: 3": "Secure Home Debt", "Unnamed: 4": "Secure Business Debt", "Unnamed: 5": "Secure Vehicle Debt", 
    "Unnamed: 7": "Unsecure Credit Card Debt", "Unnamed: 8": "Unsecure Loans", "Unnamed: 9": "Unsecure Other Debt"})).reset_index(drop=True)
secure_unsecure_debt_region
```

```python
final_df = pd.merge(secure_unsecure_debt_region, region_df, on=['Characteristic'])
final_df
```

```python
file_name2 = 'region_median_of_debt2011.xlsx'
final_df.to_excel(file_name1)
```
