### Age wise Wealth in 2011 

```python
agewise_median_wealth = pd.DataFrame(data.loc[4:10].rename(index=str, columns={'Characteristic':'Age groups', 
                                         'Unnamed: 4':'Interest earning',
                                         'Unnamed: 5': 'Regular check-in accounts',
                                         'Unnamed: 14':'IRA or KEOGH Accounts',
                                         'Unnamed: 15':'401K & Thrift Savings Plan'})).reset_index(drop=True)
agewise_median_wealth
```

```python
file_name = 'agegroup_median_wealth_2013.xlsx'
agewise_median_wealth.to_excel(file_name)
```
